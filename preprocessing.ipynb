```python


# Create list of punctuation 
punctuations = string.punctuation

# Create list of stopwords
nlp = spacy.load('en_core_web_sm')
stop_words = spacy.lang.en.stop_words.STOP_WORDS
# Load English tokenizer, tagger, parser, NER and word vectors
parser = English()

# Creating our clean function to make preprocessing
def clean(sentence):
    # Creating  token object
    mytokens = parser(sentence)

    # Lemmatizing each token and converting each token into lowercase and delete spaces and pronouns
    mytokens = [ word.strip().lemma_.lower().token() if word.lemma_ != "-PRON-"  else word.lower_ for word in mytokens ]

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens
    
```

```python
X, Y = df.verified_reviews, df.feedback

#delete stopwords and most frequent words and less used words
tfidfconverter  = TfidfVectorizer( min_df=2, max_df=0.9, stop_words=stopwords.words('english'))
X_Vec = tfidfconverter.fit_transform(X)
X_Vec = pd.DataFrame.sparse.from_spmatrix(X_Vec)

print(f'The old shape for X is {X.shape}')
print(f'The new shape for X is {X_Vec.shape}')
X_Vec.head()
```
